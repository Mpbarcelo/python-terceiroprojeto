```python
idades = []

alturas = []

for i in range(5):

    print(f"Pessoa {i + 1}:")

    idade = int(input("Digite a idade: "))

    altura = float(input("Digite a altura (em metros): "))

    idades.append(idade)

    alturas.append(altura)

print("\nIdades:")

for idade in reversed(idades):

    print(idade)

print("\nAlturas:")

for altura in reversed(alturas):

    print(f"{altura:.2f} metros")
```

```python
# Criando uma lista heterogênea com diferentes tipos de dados

dados_pessoa = ["João", 25, 1.75, True]

# Acessando elementos da lista

nome = dados_pessoa[0]

idade = dados_pessoa[1]

altura = dados_pessoa[2] * 2

e_maior = dados_pessoa[0]

# Exibindo informações

print("Nome:", nome)

print("Idade:", idade)

print("Altura:", altura)

print("É maior de idade?", e_maior)
```

```python

pessoas = []

for i in range(4):

    print(f"Cadastro da pessoa {i + 1}:")

    nome = input("Nome: ")

    idade = int(input("Idade: "))

    cpf = input("CPF: ")

    pessoa = {"Nome": nome, "Idade": idade, "CPF": cpf}

    pessoas.append(pessoa)

pessoas_2 = []

for pessoa in pessoas:

    if pessoa["Idade"] < 18:

        pessoas_2.append(pessoa)

        pessoas.remove(pessoa)

print("\nPessoas 1:")

for pessoa in pessoas:

    print(pessoa)

print("\nPessoas 2:")

for pessoa in pessoas_2:

    print(pessoa)
```
